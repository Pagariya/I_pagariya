1. Cropping image, resizing it and subtracting two images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'path1')
#print(img.shape) # Print image shape
img2 = cv2.imread(r'path2')
#print(img2.shape)
scale_percent = 20      # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
#Cropping an image
resized1 = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
resized2 = cv2.resize(img2, dim, interpolation = cv2.INTER_AREA)
#cv2.imshow("Resized image", resized)


#print(resized.shape)

cropped_image1 = resized1[ 320:380, 380:445]
cropped_image2 = resized2[ 320:380, 380:445]

# cv2.imshow("cropped", cropped_image2)


subtracted = cv2.subtract(cropped_image2, cropped_image1)


scale_percent = 800    # percent of original size
width = int(subtracted.shape[1] * scale_percent / 100)
height = int(subtracted.shape[0] * scale_percent / 100)
dim = (width, height)


 

resized3 = cv2.resize(subtracted, dim, interpolation = cv2.INTER_AREA)
fig, axes = plt.subplots(1,3)
cropped_image1 = cv2.cvtColor(cropped_image1, cv2.COLOR_BGR2RGB)
cropped_image2 = cv2.cvtColor(cropped_image2, cv2.COLOR_BGR2RGB)
resized3 = cv2.cvtColor(resized3, cv2.COLOR_BGR2RGB)
axes[0].imshow(cropped_image2)
axes[1].imshow(cropped_image1)
axes[2].imshow(resized3)
fig.set_figwidth(50)
fig.set_figheight(50)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
